```python
import os
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import timm
import torch.nn as nn
import torch.optim as optim
import gc
```

```python
BATCH_SIZE = 16   # Angka aman untuk VRAM 4GB
IMG_SIZE = 224    # Ukuran wajib input ViT
NUM_WORKERS = 0   # Set 0 agar stabil di Windows
DATA_PATH = './data'

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"ðŸ–¥ï¸ Hardware: {device}")
if torch.cuda.is_available():
    print(f"   GPU Name: {torch.cuda.get_device_name(0)}")
    # Bersihkan memori sisa sebelumnya (Penting!)
    torch.cuda.empty_cache()
    gc.collect()
```

## Load Dataset

```python
# Resize gambar ke 224x224 dan normalisasi
transform_train = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

transform_test = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load Dataset dari folder lokal
# 'download=True' akan mengecek folder ./data. Karena file sudah ada, dia akan extract otomatis.
train_set = torchvision.datasets.CIFAR10(root=DATA_PATH, train=True,
                                        download=True, transform=transform_train)
train_loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS)

test_set = torchvision.datasets.CIFAR10(root=DATA_PATH, train=False,
                                       download=True, transform=transform_test)
test_loader = DataLoader(test_set, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS)

class_names = train_set.classes
```

## Setup Model ViT & Swin

```python
def setup_model(model_name, label_name):
    print(f"   Creating: {label_name} ({model_name})...")
    
    
    model = timm.create_model(model_name, pretrained=True, num_classes=10)
        
    model = model.to(device)
    
    # Info Parameter
    total_params = sum(p.numel() for p in model.parameters())
    size_mb = total_params * 4 / (1024**2)
    print(f"     -> Params: {total_params:,} | Size: {size_mb:.2f} MB")
    
    return model


# load 
vit_model = setup_model('vit_tiny_patch16_224', "ViT-Tiny")
swin_model = setup_model('swin_tiny_patch4_window7_224', "Swin-Tiny")


optimizer_vit = optim.AdamW(vit_model.parameters(), lr=5e-5)
optimizer_swin = optim.AdamW(swin_model.parameters(), lr=5e-5)
criterion = nn.CrossEntropyLoss()
```

## Train Model

```python
import time
import copy

# Fungsi Training Utama
def train_model(model, train_loader, val_loader, criterion, optimizer, model_name, num_epochs=10):
    print(f"\n{'='*30}")
    print(f"ðŸš€ START TRAINING: {model_name}")
    print(f"{'='*30}")
    
    start_time = time.time()
    
    # Dictionary untuk menyimpan data history (Wajib untuk Grafik Laporan) [cite: 44-45]
    history = {
        'train_loss': [], 'train_acc': [],
        'val_loss': [], 'val_acc': []
    }
    
    best_model_wts = copy.deepcopy(model.state_dict())
    best_acc = 0.0
    
    for epoch in range(num_epochs):
        print(f'\nEpoch {epoch+1}/{num_epochs}')
        print('-' * 20)

        # --- FASE 1: TRAINING (Belajar) ---
        model.train() 
        running_loss = 0.0
        running_corrects = 0
        
        # Loop per batch
        for batch_idx, (inputs, labels) in enumerate(train_loader):
            inputs = inputs.to(device)
            labels = labels.to(device)

            optimizer.zero_grad() # Reset gradien

            # Forward -> Backward -> Optimize
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            # Hitung statistik batch
            _, preds = torch.max(outputs, 1)
            running_loss += loss.item() * inputs.size(0)
            running_corrects += torch.sum(preds == labels.data)
            
            # Tampilkan progress setiap 500 batch (supaya tidak hening)
            if (batch_idx + 1) % 500 == 0:
                print(f"   > Batch {batch_idx + 1}/{len(train_loader)} processed...")

        epoch_loss = running_loss / len(train_loader.dataset)
        epoch_acc = running_corrects.double() / len(train_loader.dataset)

        # --- FASE 2: VALIDATION (Ujian) ---
        model.eval()
        val_running_loss = 0.0
        val_running_corrects = 0
        
        with torch.no_grad(): # Hemat memori GPU
            for inputs, labels in val_loader:
                inputs = inputs.to(device)
                labels = labels.to(device)

                outputs = model(inputs)
                loss = criterion(outputs, labels)
                _, preds = torch.max(outputs, 1)

                val_running_loss += loss.item() * inputs.size(0)
                val_running_corrects += torch.sum(preds == labels.data)
        
        val_loss = val_running_loss / len(val_loader.dataset)
        val_acc = val_running_corrects.double() / len(val_loader.dataset)

        # Simpan history
        history['train_loss'].append(epoch_loss)
        history['train_acc'].append(epoch_acc.item())
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_acc.item())

        print(f'   Train Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
        print(f'   Val Loss:   {val_loss:.4f} Acc: {val_acc:.4f}')

        # Simpan model terbaik
        if val_acc > best_acc:
            best_acc = val_acc
            best_model_wts = copy.deepcopy(model.state_dict())
            # Simpan file fisik (Opsional untuk dikumpulkan)
            torch.save(model.state_dict(), f'{model_name}_best.pth')
            print(f"   âœ… Model terbaik disimpan (Acc: {best_acc:.4f})")

    time_elapsed = time.time() - start_time
    print(f'\nTraining selesai dalam {time_elapsed // 60:.0f}m {time_elapsed % 60:.0f}s')
    print(f'Best Val Acc: {best_acc:.4f}')
    
    # Load bobot terbaik sebelum kembali
    model.load_state_dict(best_model_wts)
    return history
```

```python
EPOCHS = 10

# 1. Latih ViT-Tiny
# Bersihkan cache dulu
if torch.cuda.is_available(): torch.cuda.empty_cache() 
vit_history = train_model(vit_model, train_loader, test_loader, criterion, optimizer_vit, "ViT-Tiny", num_epochs=EPOCHS)

# 2. Latih Swin-Tiny
print("\nIstirahat sejenak membersihkan memori GPU...")
if torch.cuda.is_available(): 
    torch.cuda.empty_cache()
    gc.collect()
    time.sleep(2)

swin_history = train_model(swin_model, train_loader, test_loader, criterion, optimizer_swin, "Swin-Tiny", num_epochs=EPOCHS)
```

## Evaluasi dan Visualisasi

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import time
```

```python
# Plot Kurva
def plot_training_history(history, model_name):
    
    acc = history['train_acc']
    val_acc = history['val_acc']
    loss = history['train_loss']
    val_loss = history['val_loss']
    epochs_range = range(1, len(acc) + 1)

    plt.figure(figsize=(12, 4))
    
    # Plot Akurasi
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, acc, label='Training Accuracy')
    plt.plot(epochs_range, val_acc, label='Validation Accuracy')
    plt.title(f'{model_name} - Accuracy')
    plt.legend(loc='lower right')
    plt.grid(True)

    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, loss, label='Training Loss')
    plt.plot(epochs_range, val_loss, label='Validation Loss')
    plt.title(f'{model_name} - Loss')
    plt.legend(loc='upper right')
    plt.grid(True)
    
    plt.tight_layout()
    plt.show()
```

```python
def evaluate_model_comprehensive(model, dataloader, model_name):
    print(f"\nðŸ“Š --- HASIL EVALUASI: {model_name} ---")
    model.eval()
    
    all_preds = []
    all_labels = []
    
    # Pengukuran Waktu 
    print("   Melakukan warm-up GPU/CPU...")
    dummy_input = torch.randn(1, 3, 224, 224).to(device)
    for _ in range(10):
        _ = model(dummy_input)
        
    # Ukur Waktu Nyata
    print("   Mulai pengukuran waktu inferensi pada Test Set...")
    start_time = time.time()
    
    with torch.no_grad():
        for inputs, labels in dataloader:
            inputs = inputs.to(device)
            labels = labels.to(device)
            
            outputs = model(inputs)
            
            # Pastikan dimensi benar
            if outputs.dim() > 2: outputs = outputs.view(outputs.size(0), -1)
                
            _, preds = torch.max(outputs, 1)
            
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            
    end_time = time.time()
    
    # Hitung Metrik Waktu
    total_time = end_time - start_time
    num_images = len(dataloader.dataset)
    avg_time_ms = (total_time / num_images) * 1000
    throughput = num_images / total_time
    
    print(f"Total Waktu: {total_time:.2f} detik")
    print(f"Rata-rata per gambar: {avg_time_ms:.2f} ms")
    print(f"Throughput: {throughput:.2f} images/sec")
    
    # B. Classification Report (Precision, Recall, F1) [cite: 36-39]
    print("\nLaporan Klasifikasi:")
    print(classification_report(all_labels, all_preds, target_names=class_names))
    
    # C. Confusion Matrix [cite: 42]
    cm = confusion_matrix(all_labels, all_preds)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.title(f'Confusion Matrix - {model_name}')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()
    
    return avg_time_ms, throughput
```

```python
# Visualisasi Kurva Training
try:
    plot_training_history(vit_history, "ViT-Tiny")
    plot_training_history(swin_history, "Swin-Tiny")
except NameError:
    print("History tidak ditemukan. Pastikan Tahap 2 sudah dijalankan sampai selesai.")

# Evaluasi Metrik & Waktu
print("\n" + "="*40)
print("MENGHITUNG PERFORMA & INFERENSI")
print("="*40)

# Evaluasi ViT
vit_time, vit_fps = evaluate_model_comprehensive(vit_model, test_loader, "ViT-Tiny")

# Evaluasi Swin
swin_time, swin_fps = evaluate_model_comprehensive(swin_model, test_loader, "Swin-Tiny")
```

```python
import pandas as pd
import torch

def save_predictions_to_csv(model, loader, model_name, class_names):
    print(f"Membuat file prediksi untuk {model_name}...")
    
    model.eval()
    results = []
    index_counter = 0
    
    with torch.no_grad():
        for inputs, _ in loader: # Kita tidak butuh label asli, cuma butuh input gambar
            inputs = inputs.to(device)
            outputs = model(inputs)
            
            # Ambil kelas dengan probabilitas tertinggi
            _, preds = torch.max(outputs, 1)
            
            # Simpan hasil batch ini
            for pred in preds:
                class_name = class_names[pred.item()]
                results.append({
                    "Image_ID": index_counter,
                    "Predicted_Label": class_name
                })
                index_counter += 1
    
    # Simpan ke CSV
    filename = f"prediksi_{model_name}.csv"
    df = pd.DataFrame(results)
    df.to_csv(filename, index=False)
    print(f"File tersimpan: {filename}")


# Buat CSV untuk ViT
save_predictions_to_csv(vit_model, test_loader, "ViT-Tiny", class_names)

# Buat CSV untuk Swin 
save_predictions_to_csv(swin_model, test_loader, "Swin-Tiny", class_names)
```
